## Exercise 4: Validation curve and Learning curve

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import ValidationCurveDisplay, learning_curve
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold
import time
```

```python
X, y = make_classification(n_samples=100000, n_features= 30, n_informative=10, flip_y=0.2)

plt.figure(figsize=(8, 6))
disp = ValidationCurveDisplay.from_estimator(
    RandomForestClassifier(),
    X,
    y,
    param_name='max_depth',
    param_range = np.arange(1, 21),
    n_jobs=-1,
    cv=5,
    scoring="roc_auc",
    score_name="ROC AUC",
)
disp.ax_.set_title("Validation curve: max_depth")
disp.ax_.set_xlabel("max_depth")
disp.ax_.set_ylabel("Score: ROC AUC")

plt.show()
```

```python
#2)

clf = RandomForestClassifier(max_depth=12)

# Define a 10-fold cross-validation strategy
cv = KFold(n_splits=10)

# Plot learning curve
def plot_learning_curve(estimator, X, y, cv, n_jobs):
    train_sizes, train_scores, test_scores, fit_times, _ = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, return_times=True, 
        train_sizes=np.linspace(0.1, 1.0, 10)
    )
    
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    fit_times_mean = np.mean(fit_times, axis=1)
    fit_times_std = np.std(fit_times, axis=1)

    # Learning curve
    plt.figure(figsize=(18, 5))

    plt.subplot(1, 3, 1)
    plt.title("Learning Curves of Random Forest")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g", label="Cross-validation score")
    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1, color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    plt.legend(loc="best")

    # Scalability of the model
    plt.subplot(1, 3, 2)
    plt.title("Scalability of the Model")
    plt.plot(train_sizes, fit_times_mean, 'o-')
    plt.fill_between(train_sizes, fit_times_mean - fit_times_std,
                     fit_times_mean + fit_times_std, alpha=0.1)
    plt.xlabel("Training examples")
    plt.ylabel("Fit time")

    # Performance of the model
    plt.subplot(1, 3, 3)
    plt.title("Performance of the Model")
    plt.plot(fit_times_mean, test_scores_mean, 'o-')
    plt.fill_between(fit_times_mean, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1)
    plt.xlabel("Fit times")
    plt.ylabel("Score")

    plt.tight_layout()
    plt.show()

# Run the function and plot
plot_learning_curve(clf, X, y, cv=cv, n_jobs=-1)
```
